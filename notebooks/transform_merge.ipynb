# Merge the Transformed Data Sets

```python
import pandas as pd
```

Let's make sure we're confident of the results by using the results of the transforming saved into CSV files to work on merging them!

```python
# Import the cleaned datasets from CSV files into DataFrames
transactions = pd.read_csv(
    "../tests/test_data/expected_transactions_clean_results.csv"
)

customers = pd.read_csv(
    "../tests/test_data/expected_customers_clean_results.csv"
)
```

The let's perform the merge based on the `customer_id`

```python
merged = pd.merge(transactions, customers, on="customer_id")

# View the first few rows of the merged DataFrame
merged.head()
```

There should be no duplicates - but let's check

```python
duplicates = merged[merged.duplicated()]

duplicates
```

And now we're happy, lets export the result of the merge into a CSV file so we can test against it in the pipeline (and use it in our enrich/aggregation step)

```python
merged.to_csv(
    "../tests/test_data/expected_merged_clean_results.csv", index=False
)
```
