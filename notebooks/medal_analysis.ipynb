# Streamlit Visualisations

```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt

file_path = "../data/processed/transformed_data.csv"

df = pd.read_csv(file_path)
```

```python
#  Summer and Winter bar chart of top 10 countries by total medals

summer = df[df["season"] == "Summer"]
winter = df[df["season"] == "Winter"]

summer_medal_count = summer.groupby("country")["medal"].count().sort_values(ascending=False)
winter_medal_count = winter.groupby("country")["medal"].count().sort_values(ascending=False)
summer_gold_medal_count = summer[summer["medal"] == "Gold"].groupby("country")["medal"].count().sort_values(ascending=False)
winter_gold_medal_count = winter[winter["medal"] == "Gold"].groupby("country")["medal"].count().sort_values(ascending=False)
```
